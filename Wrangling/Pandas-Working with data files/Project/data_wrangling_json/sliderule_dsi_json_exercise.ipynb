## JSON example, with string

+ demonstrates creation of normalized dataframes (tables) from nested json string
+ source: http://pandas.pydata.org/pandas-docs/stable/io.html#normalization

```python
# define json string
data = [{'state': 'Florida', 
         'shortname': 'FL',
         'info': {'governor': 'Rick Scott'},
         'counties': [{'name': 'Dade', 'population': 12345},
                      {'name': 'Broward', 'population': 40000},
                      {'name': 'Palm Beach', 'population': 60000}]},
        {'state': 'Ohio',
         'shortname': 'OH',
         'info': {'governor': 'John Kasich'},
         'counties': [{'name': 'Summit', 'population': 1234},
                      {'name': 'Cuyahoga', 'population': 1337}]}]
```

```python
# use normalization to create tables from nested element
json_normalize(data, 'counties')
```

```python
json_normalize(data, 'counties', ['state', 'shortname'])
```

```python
# further populate tables created from nested element
json_normalize(data, 'counties', ['state', 'shortname', ['info', 'governor']])
```

## Q1: Find the 10 countries with most projects

#### Load database

```python
wb_proj_df = pd.read_json('data/world_bank_projects.json')
wb_proj_df.head()
```

#### arranging database by groupping based on countries, finding number of projects done in each country

```python
proj_country_df = wb_proj_df.groupby(['countryshortname']).size().sort_values(ascending=False).reset_index()
proj_country_df = proj_country_df.rename(columns={0 : 'Number of Projects'})
proj_country_df.head()
```

#### rank top 10 countries with most projects

```python
proj_country_df['rank'] = proj_country_df['Number of Projects'].rank(ascending=0, method='min')
proj_country_df = proj_country_df[proj_country_df['rank']<=10]
proj_country_df.set_index('rank', inplace=True)
proj_country_df
```

## Q2: Find the top 10 major project themes (using column 'mjtheme_namecode')

***
#### First, reading only needed part of database ('mjtheme_namecode', 'countryshortname', 'id') using json Normalization 

```python
data = json.load((open('data/world_bank_projects.json')))
mjtheme_df = json_normalize(data, 'mjtheme_namecode', ['countryshortname', ['_id', '$oid']])
mjtheme_df = mjtheme_df.rename(columns={'_id.$oid':'id'})
mjtheme_df.head()
```

#### Grouping table based on 'id', now we have 'name' and 'code' of each major theme of projects

```python
grep = mjtheme_df.groupby(['id'])
```

#### Each project has multiple major themes, no we combine the codes and names of each project.

```python
code_df = grep.code.apply(lambda x:','.join(x.astype(str))).reset_index()
code_df.head()
```

```python
name_df = grep['name'].apply(lambda x: '{%s }' % ', '.join(x)).reset_index()
name_df.head()
```

```python
theme_namecode_df = pd.merge(name_df, code_df, on='id')
theme_namecode_df.head()
```

#### now, this is time find top 10 major themes of projects by groupping table above based on 'code'

```python
theme_rank = theme_namecode_df.groupby(['code', 'name']).size().sort_values(ascending=False).reset_index()
theme_rank = theme_rank.rename(columns={0:'Number of Projects'})
theme_rank['rank'] = theme_rank['Number of Projects'].rank(ascending=0, method='min')
top_mjtheme = theme_rank[theme_rank['rank']<=10].set_index('rank')
top_mjtheme
```

## Q3: In 2. above you will notice that some entries have only the code and the name is missing. Create a dataframe with the missing names filled in.

#### Reading needed part of database, such as done in Q2., and also, convert missed values of names to NaN

```python
mjtheme_mod = json_normalize(data, 'mjtheme_namecode', ['countryshortname', ['_id', '$oid']])
mjtheme_mod = mjtheme_mod.rename(columns={'_id.$oid':'id', 'countryshortname':'country'})
mjtheme_mod['name'][mjtheme_mod['name']==''] = np.nan
mjtheme_mod.head()
```

#### as seen there is some NaN value in name column. sometimes we have no choice but dropping out these rows. Fortunarely, here we can restore these missing data from others. For example, in the table above, the name of the second row is missed, but as we now its code (11), we can find the corresponding name for code=11, and fill in the missing cell.

#### So, first we prepare a mapping code-to-name table.

```python
code_map_mod = mjtheme_mod[['code', 'name']].set_index('code').dropna().drop_duplicates().reset_index()
code_map_mod
```

#### Simply merging the map table with original table, the missing values are filled. in the below table you can see the name column of old table and the refined new table. as seen, the missing values are now filled with a name mapped from its code.

```python
mjtheme_mod = pd.merge(mjtheme_mod, code_map_mod, on='code', suffixes=['_old', '_new'])
mjtheme_mod.head()
```

#### Now, we redo the process in Q2, again:

```python
top_mjtheme_mod = mjtheme_mod[['id', 'code', 'name_new', 'country']].sort_values(by='id')
top_mjtheme_mod = top_mjtheme_mod.rename(columns={'name_new':'name'})
top_mjtheme_mod.head()
```

```python
grep_mod = top_mjtheme_mod.groupby('id')
code_df_mod = grep_mod.code.apply(lambda x:', '.join(x.astype(str))).reset_index()
name_df_mod = grep_mod['name'].apply(lambda x:', '.join(x)).reset_index()
theme_namecode_df_mod = pd.merge(name_df_mod, code_df_mod, on='id')
theme_namecode_df_mod.head()
```

```python
theme_rank_mod = theme_namecode_df_mod.groupby(['code', 'name']).size().sort_values(ascending=False).reset_index()
theme_rank_mod = theme_rank_mod.rename(columns={0:'Number of Projects'})
theme_rank_mod['rank'] = theme_rank_mod['Number of Projects'].rank(ascending=0, method='min')
top_mjtheme_mod = theme_rank_mod[theme_rank_mod['rank']<=10].set_index('rank')
top_mjtheme_mod
```
